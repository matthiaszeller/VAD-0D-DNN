```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = '/media/maousi/Data/tmp/tmp/simulation_LVAD_Art_Pulse_RPM5000_2020_05_31/results/'
```

```python
columns = [
    'HR','SAPM','SAPS','SAPD', 'PAPM','PAPS','PAPD','LVEF',
    'LVEDV', 'LVESV', 'CI', 'PCPW'
]
```

```python
X_ex = pd.read_csv(path + 'Xexact.csv', header=None)
X_pred = pd.read_csv(path + 'Xpredicted.csv', header=None)
X_ex.columns = columns
X_pred.columns = columns
```

```python
X = X_ex.join(X_pred, rsuffix='_pred')
```

```python
X.head()
```

```python
fig, ax = plt.subplots(4, 3, figsize=(10,10))
ax = ax.ravel()

for i, col in enumerate(columns):
    ax[i].boxplot(X[col])
    ax[i].set_ylabel(col)

    plt.subplots_adjust(wspace=.4, hspace=.2)
```
